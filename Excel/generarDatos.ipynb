```python
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
import random

# Configurar parámetros
regiones = ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Piura']
productos = {
    'Arroz': 'Alimentos',
    'Aceite': 'Alimentos',
    'Galletas': 'Snacks',
    'Jugo': 'Bebidas',
    'Cerveza': 'Bebidas',
    'Papel Higiénico': 'Limpieza',
    'Shampoo': 'Limpieza',
    'Leche': 'Lácteos',
    'Yogurt': 'Lácteos',
    'Queso': 'Lácteos'
}
canales = ['Tienda', 'Online', 'Mayorista']

# Carpeta destino
output_folder = "VentasRegionales"
os.makedirs(output_folder, exist_ok=True)

# Función para generar un archivo por región
def generar_datos_region(region):
    num_filas = 100  # Puedes aumentar esto para más datos
    fechas = [datetime(2024, 1, 1) + timedelta(days=random.randint(0, 180)) for _ in range(num_filas)]
    productos_lista = random.choices(list(productos.keys()), k=num_filas)
    categorias = [productos[p] for p in productos_lista]
    canales_lista = random.choices(canales, k=num_filas)
    unidades = [random.randint(1, 20) for _ in range(num_filas)]
    precios_unitarios = [random.uniform(2.0, 15.0) for _ in range(num_filas)]
    ventas = [round(unidades[i] * precios_unitarios[i], 2) for i in range(num_filas)]

    df = pd.DataFrame({
        'Fecha': fechas,
        'Región': [region]*num_filas,
        'Producto': productos_lista,
        'Categoría': categorias,
        'Canal': canales_lista,
        'Unidades': unidades,
        'Ventas': ventas
    })

    ruta = os.path.join(output_folder, f"Ventas_{region}.xlsx")
    df.to_excel(ruta, index=False)
    print(f"Archivo generado: {ruta}")

# Generar archivos para todas las regiones
for region in regiones:
    generar_datos_region(region)
```
